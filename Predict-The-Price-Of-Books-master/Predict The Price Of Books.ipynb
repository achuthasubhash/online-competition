```python
bookcat =pd.get_dummies(train_df.BookCategory,drop_first=True)
```

```python
train_df = pd.concat([train_df,bookcat],axis=1)
```

```python
X=train_df[['review','rating','Arts, Film & Photography','Biographies, Diaries & True Accounts','Comics & Mangas','Computing, Internet & Digital Media','Crime, Thriller & Mystery','Humour','Language, Linguistics & Writing','Politics','Romance','Sports']]
```

```python
X
```

```python
y=train_df.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor1 = RandomForestRegressor(n_estimators = 10, random_state = 0) # as n inc,acc inc
regressor1.fit(X_train, y_train)
```

```python
y_pred = regressor1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
print(y_test, y_pred)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
x=np.arange(0,1248)
```

```python
plt.scatter(x, y_test, color = 'red')
plt.plot(x, regressor1.predict(X_test), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
