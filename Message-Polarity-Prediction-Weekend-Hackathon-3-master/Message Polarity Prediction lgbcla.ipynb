```python
testdata=pd.read_csv("E:\\machine hack\\Message Polarity Prediction\\MessagePolarity_ParticipantsData\\Test.csv")
```

```python
fdata=traindata.append(testdata,sort=False)
```

```python
fdata.reset_index(inplace=True)
```

```python
y=traindata.IsGoodNews
```

```python
fdata.drop(['IsGoodNews'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_sc = scaler.fit_transform(fdata)
```

```python
df = pd.DataFrame(data=X_sc)
```

```python
df
```

```python
X_train=df[0:947]
y_train=y
X_test=df[947:]
```

```python
from lightgbm import LGBMClassifier
classifier = LGBMClassifier( random_state=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
import xlsxwriter 
submission=pd.DataFrame()
submission['IsGoodNews']=y_pred
submission.to_csv('lgb classifier.csv',index=False)
```
