```python

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_sc = sc_X.fit_transform(x)
```

```python
y=data['Infect_Prob']
```

```python
data=pd.read_excel("E:\\assignment\\ass 2\\Test_dataset.xlsx")
data.isnull().values.any()

X=data
X.head()
gen=pd.get_dummies(data['Gender'],drop_first=True)
tra=pd.get_dummies(data['Mode_transport'],drop_first=True)
data['comorbidity'].unique()
com=pd.get_dummies(data['comorbidity'],drop_first=True)
data.drop(['Region','Gender','Mode_transport','comorbidity'],axis=1,inplace=True)
data = pd.concat([data,tra,gen,com],axis=1)
fea=['Children','cases/1M','Deaths/1M','Age','Coma score','Diuresis','Platelets','HBB','d-dimer','Heart rate','HDL cholesterol','Charlson Index','Blood Glucose','Insurance','FT/month','Public','Walk','Male','Diabetes','Hypertension','None']
x_test=data[fea]
y_pred = regressor.predict(x_test)
```
